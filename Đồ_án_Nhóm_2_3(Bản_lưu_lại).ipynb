Bằng cách dự báo chính xác, các nhà hoạch định có thể tối ưu hóa nguồn cung, giảm lãng phí và thúc đẩy sử dụng năng lượng tái tạo, góp phần bảo vệ môi trường và nâng cao hiệu quả kinh tế.

  - Phân tích và dự báo nhu cầu tiêu thụ điện năng đóng vai trò quan trọng trong việc đảm bảo cân bằng cung - cầu, hỗ trợ phát triển kinh tế - xã hội, tối ưu hóa chi phí đầu tư và vận hành, cũng như thúc đẩy sử dụng năng lượng tái tạo.   - Phân tích thăm dò dữ liệu (Exploratory Data Analysis - EDA) là bước quan trọng nhằm hiểu rõ đặc điểm, cấu trúc và xu hướng của dữ liệu tiêu thụ điện năng. Phương pháp này bao gồm thống kê mô tả, xác định các phân phối dữ liệu, kiểm tra tính bất thường và phát hiện xu hướng tiềm ẩn.

  - Các phương pháp chính để phân tích thăm dò và trực quan dữ liệu về nhu cầu sử dụng điện năng bao gồm:
     + Thống kê mô tả: Tính toán các giá trị trung bình, trung vị, phương sai, độ lệch chuẩn để hiểu rõ đặc điểm dữ liệu.

     + Phân tích thành phần chính (PCA): Giảm số chiều dữ liệu để tìm ra các yếu tố quan trọng nhất ảnh hưởng đến tiêu thụ điện năng.

     + Học máy (Machine Learning): Áp dụng mô hình hồi quy tuyến tính, cây quyết định, mạng nơ-ron nhân tạo (ANN) để dự báo nhu cầu tiêu thụ điện chính xác hơn.

## 3.Thống kê mô tả
- Các đại lượng thống kê cơ bản :
     + Tần suất (Frequency): Số lần xuất hiện của từng mức tiêu thụ điện trong khoảng khác nhau (trên 1000 TWh, 500-1000 TWh, dưới 500 TWh, v.v.).
     + Trung bình (Mean): Trung bình tiêu thụ điện của các quốc gia đang xét.
     + Trung vị (Median): Gía trị nằm giữa tập dữ liệu khi sắp xếp theo thứ tự tang dần.
     + Phương sai (Variance) và Độ lệch chuẩn (Standard Deviation): Đo độ phân tán của mức độ tiêu thụ điện quanh giá trị trung bình.
     + Phân phối (Distribution): Ước lượng tỷ lệ giúp xác định phần tram trong nhóm tổng thể.
- Phân tích phân phối dữ liệu: Để đánh giá mức độ tiêu thụ điện năng của các quốc gia được chọn:
     + Sử dụng ước lượng tỷ lệ: Xác định xác suất đạt từng mức sử dụng điện, giúp dự đoán nhu cầu sử dụng điện của các nước
     + Sử dụng ước lượng hồi quy tuyến tính: Dự đoán mức độ tiêu thụ điện dựa trên dân số.
     + Thực hiện kiểm định Anova: Xác định xem có sự khác biệt đáng kể giữa các quốc gia hay không.
     + Sử dụng kiểm định Shapiro-Wilk (Shapiro-Wilk Test): Kiểm tra xem dữ liệu có tuân theo phân phối chuẩn không.
     + Thực hiện kiểm định Chi-square: Kiểm tra sự liên hệ giữa hai biến phân loại.
      
- Biểu đồ tóm tắt dữ liệu:
     + Sử dụng biểu đồ (Boxplot): Hiện thị sự phân bố của dữ liệu, phát hiện các giá trị ngoại lai. So sánh mức tiêu thụ điện giữa các giốc gia và khu vực.
      + Sử dụng biểu đồ cột (Bar plot): So sánh trực tiếp giá trị của các quốc gia, giúp dễ dàng thấy quốc gia nào tiêu thụ nhiều nhất.
      + Sử dụng biểu đồ đường (Line plot): Theo dõi xu hướng theo thời gian hoặc chỉ một chỉ số liên tục.
      + Sử dụng biểu đồ Histogram: Kiêm tra phân bố của dữ liệu, xem có tập trung và phân tán vào một khoảng hay không.
- Sử dụng ước lượng mô hình:
     + Mô hình ARDL (Autoregressive Distributed Lag Model): Đánh giá mối quan hệ dài hạn và ngắn hạn giữa tiêu thụ điện và các yếu tố kinh tế.
- Sử dụng kiểm định thống kê:
     + Kiểm định tương quan Pearson/Spearman: Đánh giá mức độ liên hệ giữa tiêu thụ điện và từng yếu tố.
     + Kiểm định nhân quả Granger: Xác định yếu tố nào có thể là nguyên nhân dẫn đến thay đổi tiêu thụ điện năng.
     + Kiểm định tính dừng (ADF, KPSS): Kiểm tra tính ổn định của dữ liệu chuỗi thời gian trước khi hồi quy.
     + Kiểm định phương sai thay đổi (Breusch-Pagan, White Test): Kiểm tra tính đồng nhất của phương sai trong mô hình hồi quy.
     + Kiểm định đồng liên kết (Johansen, Engle-Granger Test): Kiểm tra quan hệ dài hạn giữa tiêu thụ điện và các yếu tố khác.



## 4.Phân tích mối liên hệ giữa các biến
- Biểu đồ phân tán (Scatter Plot): Dùng để hiển thị mối quan hệ giữa hai biến liên tục, ví dụ:
     + Tiêu thụ điện năng & GDP
     + Tiêu thụ điện năng & nhiệt độ trung bình
     + Tiêu thụ điện năng & dân số
     + Tiêu thụ điện năng & lượng khí thải CO₂
     + GDP & dân số

- Ma trận tương quan (Correlation Matrix): Hiển thị hệ số tương quan giữa nhiều biến liên tục cùng lúc, giúp nhận diện mối quan hệ mạnh/yếu giữa các yếu tố như:
     + Tiêu thụ điện năng, GDP, dân số, nhiệt độ, lượng khí thải, phát triển công nghiệp, năng lượng tái tạo, v.v.

- Hệ số tương quan (Correlation Coefficient, rrr):
     + Giá trị từ −1-1−1 đến +1+1+1, thể hiện mức độ chặt chẽ và chiều của mối quan hệ giữa hai biến.
     + r>0r > 0r>0: Hai biến có mối quan hệ đồng biến (biến này tăng, biến kia tăng).
     + r<0r < 0r<0: Hai biến có mối quan hệ nghịch biến (biến này tăng, biến kia giảm).
     + ∣r∣|r|∣r∣ gần 1: Mối quan hệ mạnh, gần 0: Quan hệ yếu hoặc không có.
- Độ đồng biến (Covariance, Cov(X,Y)\text{Cov}(X,Y)Cov(X,Y)):
     + Cho biết hai biến thay đổi cùng nhau như thế nào, nhưng không chuẩn hóa về thang đo nên khó so sánh giữa các cặp biến khác nhau.
     + Giá trị dương: Hai biến có xu hướng tăng cùng nhau.
     + Giá trị âm: Khi một biến tăng, biến kia giảm.
     + Giá trị gần 0: Không có mối quan hệ rõ ràng.

## 6.Một số chủ đề nâng cao: lựa chọn một trong các chủ để nâng cao sau để phân tích dữ liệu: Phân tích thành phần chính (PCA).
a. PCA là gì?
  - Phân tích thành phần chính (PCA) là một kỹ thuật giảm chiều dữ liệu, giúp biến đổi tập dữ liệu có nhiều biến số thành một tập hợp nhỏ hơn nhưng vẫn giữ lại phần lớn thông tin quan trọng.
  - PCA giúp loại bỏ sự dư thừa trong dữ liệu bằng cách tìm ra các hướng chính (thành phần chính) mà dữ liệu biến thiên nhiều nhất.

b. Mục đích của PCA
  - Giảm chiều dữ liệu, giúp trực quan hóa dễ dàng hơn.
  - Loại bỏ nhiễu và thông tin dư thừa.
  - Tăng hiệu suất mô hình học máy bằng cách giảm số lượng đặc trưng đầu vào.
  - Giảm hiện tượng đa cộng tuyến giữa các biến số.

c. Ứng dụng của PCA
  - Nhận dạng khuôn mặt: PCA được sử dụng trong Eigenfaces để giảm số lượng đặc trưng trong nhận diện khuôn mặt.
  - Xử lý ảnh: Giảm chiều dữ liệu ảnh để tăng tốc độ xử lý.
  - Học máy: Giảm chiều dữ liệu đầu vào để tránh overfitting.
  - Tài chính: Xác định các yếu tố chính ảnh hưởng đến biến động thị trường.

d. Hạn chế của PCA
  - Không thể xử lý dữ liệu phi tuyến tính tốt.
  - Mất đi một phần thông tin khi giảm chiều dữ liệu.
  - Kết quả khó diễn giải nếu dữ liệu không có mối quan hệ tuyến tính rõ ràng.

####Sử dụng ước lượng hồi quy tuyến tính

Mô hình này giúp xác định mối quan hệ giữa mức tiêu thụ điện của các quốc gia với dân số và mức tiêu thụ bình quân.

```r
# Chuyển đổi kiểu dữ liệu nếu cần
df$Dan_so <- as.numeric(gsub(",", "", df$Dan_so))  # Xóa dấu phẩy nếu có
df$Tieu_Thu_Binh_Quan_kWh <- as.numeric(gsub(",", "", df$Tieu_Thu_Binh_Quan_kWh))
df$Muc_Tieu_Thu_kWh <- as.numeric(gsub(",", "", df$Muc_Tieu_Thu_kWh))

# Xây dựng mô hình hồi quy tuyến tính
model <- lm(Muc_Tieu_Thu_kWh ~ Dan_so + Tieu_Thu_Binh_Quan_kWh, data = df)

# Hiển thị kết quả mô hình
summary(model)

# Kiểm tra phân phối phần dư
shapiro.test(residuals(model))

# Vẽ biểu đồ dự đoán so với thực tế
library(ggplot2)
df$Predicted <- predict(model, df)

ggplot(df, aes(x = Muc_Tieu_Thu_kWh, y = Predicted)) +
  geom_point(color = "blue") +
  geom_abline(slope = 1, intercept = 0, color = "red", linetype = "dashed") +
  labs(title = "Dự đoán vs Giá trị thực tế",
       x = "Mức tiêu thụ điện thực tế (kWh)",
       y = "Mức tiêu thụ điện dự đoán (kWh)") +
  theme_minimal()
```

**Kiểm tra phân phối phần dư:**

**Shapiro-Wilk Test:**

 - p-value = 1.439e-09 → p-value rất nhỏ → Phần dư không tuân theo phân phối chuẩn, điều này có thể ảnh hưởng đến độ tin cậy của mô hình.

 - Đường chấm đỏ thể hiện đường dự đoán hoàn hảo.

 - Các điểm dữ liệu phân bố không đều, một số giá trị rất cao (outliers) có thể làm giảm độ chính xác của mô hình.

####Sử dụng kiểm định Shapiro-Wilk (Shapiro-Wilk Test)

```r
# Kiểm tra phân phối chuẩn bằng Shapiro-Wilk test
shapiro_test <- shapiro.test(df$Muc_Tieu_Thu_kWh)
print(shapiro_test)

# Tính các thống kê mô tả
mean_value <- mean(df$Muc_Tieu_Thu_kWh, na.rm = TRUE)
sd_value <- sd(df$Muc_Tieu_Thu_kWh, na.rm = TRUE)

# Vẽ histogram với đường cong phân phối chuẩn
hist(df$Muc_Tieu_Thu_kWh,
     probability = TRUE,
     main = "Phan phoi muc tieu thu đien",
     col = "lightblue",
     breaks = 30,
     xlab = "Muc tieu thu đien (kWh)")

# Vẽ đường cong phân phối chuẩn
curve(dnorm(x, mean = mean_value, sd = sd_value),
      col = "red",
      lwd = 2,
      add = TRUE)

# Hiển thị giá trị trung bình và độ lệch chuẩn
legend("topright", legend = c(paste("Mean:", round(mean_value, 2)),
                              paste("SD:", round(sd_value, 2))),
       col = "black", cex = 0.8)
```

```r
# Cài đặt gói cần thiết
install.packages("urca")  # Gói hỗ trợ kiểm định đồng liên kết
library(urca)

# Dữ liệu từ bảng
df1 <- data.frame(
  Tong_Tieu_Thu_kWh = c(5163, 234, 400, 1502, 844, 233, 1527, 246, 75, 1056, 1695, 901, 9156, 310),
  Tong_Cong_Suat_Thuy_Dien = c(27064.0, 5956.0, 11682.0, 20148.0, 15024.0, 3840.0, 16860.0, 4200.0, 2779.0, 20950.0, 32493.0, 16240.0, 43744.0, 6467.5),
  Tong_Cong_Suat_Gio = c(890.943, 0.0, 0.0, 991.003, 2.454, 322.0, 15.145, 333.0, 0.0, 1630.846, 6.285, 1.538, 820.973, 0.0)
)

# Chuyển đổi dữ liệu thành chuỗi thời gian (giả định yếu tố thời gian)
time_series_df1 <- ts(df1)

# Kiểm định Johansen (kiểm tra đồng liên kết)
johansen_test <- ca.jo(time_series_df1, type = "trace", K = 2, ecdet = "none")
summary(johansen_test)
```

  + Giúp dự báo xu hướng dài hạn và tối ưu hóa chính sách sản xuất.

###Kiểm định phương sai thay đổi (Breusch-Pagan, White Test)

**1. Breusch-Pagan Test:**

 - Phát hiện phương sai thay đổi trong mô hình hồi quy tuyến tính (heteroscedasticity).
 - Kiểm tra xem sai số của mô hình có phụ thuộc vào biến độc lập hay không.

**2. White Test:**

 - Xác định phương sai thay đổi và các dạng sai lệch khác của mô hình hồi quy.
 - Phân tích mối quan hệ phi tuyến và tương tác giữa các biến độc lập.

```r
# Cài đặt gói cần thiết
install.packages("lmtest")
library(lmtest)

# Dữ liệu từ bảng
df1 <- data.frame(
  Tong_Tieu_Thu_kWh = c(5163, 234, 400, 1502, 844, 233, 1527, 246, 75, 1056, 1695, 901, 9156, 310),
  Tong_Cong_Suat_Thuy_Dien = c(27064.0, 5956.0, 11682.0, 20148.0, 15024.0, 3840.0, 16860.0, 4200.0, 2779.0, 20950.0, 32493.0, 16240.0, 43744.0, 6467.5),
  Tong_Cong_Suat_Gio = c(890.943, 0.0, 0.0, 991.003, 2.454, 322.0, 15.145, 333.0, 0.0, 1630.846, 6.285, 1.538, 820.973, 0.0),
  Tong_Cong_Suat_Dia_Nhiet = c(5406.970, 0.0, 0.0, 8234.000, 4212.000, 84.000, 992.000, 147.000, 0.0, 1604.000, 4541.000, 95.000, 5694.687, 0.0)
)

# Mô hình hồi quy
model <- lm(Tong_Tieu_Thu_kWh ~ Tong_Cong_Suat_Thuy_Dien + Tong_Cong_Suat_Gio + Tong_Cong_Suat_Dia_Nhiet, data = df1)

# Kiểm định Breusch-Pagan
print("Kết quả kiểm định Breusch-Pagan:")
bp_test <- bptest(model)
print(bp_test)

# Kiểm định White
print("Kết quả kiểm định White:")
white_test <- bptest(model,Tong_Cong_Suat_Thuy_Dien + I(Tong_Cong_Suat_Thuy_Dien^2) +
                     Tong_Cong_Suat_Gio + I(Tong_Cong_Suat_Gio^2) +
                     Tong_Cong_Suat_Dia_Nhiet + I(Tong_Cong_Suat_Dia_Nhiet^2))
print(white_test)
```

**1. Breusch-Pagan Test:**

 - Giá trị p-value < 0.05: Có phương sai thay đổi (heteroscedasticity).

 - Giá trị p-value ≥ 0.05: Không có phương sai thay đổi.

**2. White Test:**

 - Giá trị p-value < 0.05: Có phương sai thay đổi và khả năng sai lệch mô hình.

 - Giá trị p-value ≥ 0.05: Mô hình ổn định.

**3. Ý nghĩa:**
 - Nếu có phương sai thay đổi:

  + Ước lượng từ mô hình hồi quy không hiệu quả.

  + Cần sử dụng sai số chuẩn robust hoặc các phương pháp thay thế.

 - Nếu không có phương sai thay đổi:

  + Phương sai không đổi của sai số giúp mô hình đáng tin cậy hơn.

###Kiểm định tính dừng (ADF, KPSS)

 - Kiểm định tính dừng (ADF, KPSS) kiểm tra chuỗi thời gian có dừng hay không. Nếu dữ liệu không dừng, cần biến đổi (như lấy sai phân) để đảm bảo mô hình phân tích chính xác. Nó giúp xác định đặc điểm dữ liệu và đảm bảo kết quả phân tích đáng tin cậy.

```r
# Dữ liệu mẫu (thay bằng dữ liệu của bạn)
data <- data.frame(
  CongSuat = c(100, 120, 140, 160, 180),
  TieuThu = c(80, 95, 115, 130, 150)
)

# Thống kê mô tả cơ bản
summary(data)

# Tính toán độ phân tán
sd(data$CongSuat)  # Độ lệch chuẩn của công suất
sd(data$TieuThu)   # Độ lệch chuẩn của tiêu thụ

# Phân phối
library(e1071)
skewness(data$CongSuat)  # Độ lệch phân phối
kurtosis(data$TieuThu)   # Độ nhọn

```

**1. ADF Test:**

 - Giả thuyết gốc (H0): Chuỗi thời gian có gốc đơn (không dừng).

 - Nếu giá trị p-value < 0.05, bác bỏ giả thuyết gốc → Chuỗi thời gian là dừng.

**2. KPSS Test:**

 - Giả thuyết gốc (H0): Chuỗi thời gian là dừng.

 - Nếu giá trị p-value < 0.05, bác bỏ giả thuyết gốc → Chuỗi thời gian không dừng.

**3. Ý nghĩa**

 - Khi chuỗi thời gian là dừng:

  + Có thể áp dụng các phương pháp kinh tế lượng như hồi quy tuyến tính, ARDL, hay kiểm định nhân quả Granger.

  + Dữ liệu dừng giúp đảm bảo tính ổn định và đáng tin cậy của mô hình phân tích.

 - Khi chuỗi thời gian không dừng:

  + Cần biến đổi dữ liệu, ví dụ: lấy sai phân hoặc logarit, để đạt được tính dừng trước khi thực hiện các phân tích tiếp theo.

###Kiểm định nhân quả Granger

 - Nếu p-value ≥ 0.05, không có quan hệ nhân quả Granger giữa các biến.

 - F-statistic: Đo lường mức độ phù hợp của mô hình dự báo.

**2. Ý nghĩa:**

 - Hỗ trợ ra quyết định: Nếu có mối quan hệ nhân quả rõ ràng, điều này có thể giúp điều chỉnh chính sách sản xuất và tiêu thụ năng lượng.

###Kiểm định tương quan Pearson/Spearman

**1. Tương quan Pearson:**

 - Đánh giá mối quan hệ tuyến tính giữa hai biến số (ví dụ: Tổng Tiêu Thụ và Tổng Công Suất Thủy Điện).

 - Phù hợp với dữ liệu phân phối chuẩn.

**2. Tương quan Spearman:**

 - Đánh giá mối quan hệ đơn điệu (monotonic relationship), tức là biến này tăng thì biến kia cũng tăng hoặc ngược lại.

 - Phù hợp với dữ liệu không phân phối chuẩn hoặc dữ liệu thứ hạng (ordinal).

```r
# Dữ liệu từ bảng
data <- data.frame(
  Tong_Tieu_Thu_kWh = c(5163, 234, 400, 1502, 844, 233, 1527, 246, 75, 1056, 1695, 901, 9156, 310),
  Tong_Cong_Suat_Thuy_Dien = c(27064.0, 5956.0, 11682.0, 20148.0, 15024.0, 3840.0, 16860.0, 4200.0, 2779.0, 20950.0, 32493.0, 16240.0, 43744.0, 6467.5),
  Tong_Cong_Suat_Gio = c(890.943, 0.0, 0.0, 991.003, 2.454, 322.0, 15.145, 333.0, 0.0, 1630.846, 6.285, 1.538, 820.973, 0.0),
  Tong_Cong_Suat_Dia_Nhiet = c(600.0, 200.0, 0.0, 150.0, 1.0, 300.0, 0.0, 250.0, 0.0, 200.0, 0.0, 150.0, 900.0, 0.0)
)

# Kiểm định tương quan Pearson
pearson_result <- cor(data, method = "pearson")
print("Tương quan Pearson:")
print(pearson_result)

# Kiểm định tương quan Spearman
spearman_result <- cor(data, method = "spearman")
print("Tương quan Spearman:")
print(spearman_result)
```

**1. Hệ số tương quan Pearson:**

 - Hệ số tương quan có giá trị từ -1 đến +1:

  + +1: Quan hệ tuyến tính đồng biến hoàn hảo.

  + -1: Quan hệ tuyến tính nghịch biến hoàn hảo.

  + 0 : Không có quan hệ tuyến tính.

**2. Hệ số tương quan Spearman:**

 - Đánh giá mức độ quan hệ thứ hạng (đơn điệu) giữa các biến.

 - Hệ số tương quan có ý nghĩa tương tự Pearson nhưng phù hợp cho dữ liệu không phân phối chuẩn.

###3.2.Ma trận tương quan(Correlation Matrix)

```r
df <- read.csv("https://raw.githubusercontent.com/Tuananh-zer/file_data.csv/refs/heads/main/danh_sach_tieu_thu_dien_sach.csv", header = TRUE)
# Chọn các cột cần thiết
data_numeric <- df[, c("Muc_Tieu_Thu_kWh", "Tieu_Thu_Binh_Quan_kWh",
                       "Cong_Suat_Thuy_Dien_MW", "Cong_Suat_Gio_MW",
                       "Cong_Suat_Dia_Nhiet_MW", "Dan_so")]

# Chuyển tất cả các cột sang numeric
data_numeric <- data.frame(lapply(data_numeric, function(x) as.numeric(as.character(x))))

# Tính ma trận tương quan Kendall
cor_kendall <- cor(data_numeric, method = "kendall", use = "pairwise.complete.obs")

# In kết quả
print(round(cor_kendall, 3))

```

###3.3Hệ số tương quan(Correlation Coefficient, rrr)

```r
install.packages("reshape2")
install.packages("corrplot")
# Tải thư viện
library(ggplot2)
library(reshape2)
library(corrplot)
library(readr)

# Đọc dữ liệu từ tệp CSV
df <- read.csv("https://raw.githubusercontent.com/Tuananh-zer/file_data.csv/refs/heads/main/danh_sach_tieu_thu_dien_sach.csv")
# Chuyển cột số thành numeric nếu cần
df_numeric <- df[, sapply(df, is.numeric)]

# Tính ma trận tương quan
cor_matrix <- cor(df_numeric, use = "complete.obs")

# Vẽ biểu đồ heatmap
corrplot(cor_matrix, method = "color", type = "upper",
         tl.col = "black", tl.srt = 45,
         addCoef.col = "black", number.cex = 0.7,
         col = colorRampPalette(c("blue", "white", "red"))(200),
         title = "He so tuong quan giua cac bien",
         mar=c(0,0,1,0))
```

Hệ số tương quan (r) là một chỉ số thống kê đo lường mối liên hệ tương quan giữa hai biến số. Hệ số tương quan có giá trị từ -1 đến 1. Hệ số tương quan bằng 0 (hay gần 0) có nghĩa là hai biến số không có liên hệ gì với nhau; ngược lại nếu hệ số bằng -1 hay 1 có nghĩa là hai biến số có một mối liên hệ tuyệt đối. Nếu giá trị của hệ số tương quan là âm (r < 0) có nghĩa là hai biến tương quan nghịch (biến này tăng thì biến kia giảm và ngược lại); nếu giá trị hệ số tương quan là dương (r > 0) có nghĩa là hai biến tương quan thuận (hai biến cùng tăng hoặc cùng giảm).

**1**.Tieu_Thu_Binh_Quan_kWh vs. Tieu_Thu_Binh_Quan_kWh

Hệ số tương quan = 1.00 (màu đỏ đậm)

Đây là điều đương nhiên: một biến luôn tương quan tuyệt đối với chính nó.

**2**. Tieu_Thu_Binh_Quan_kWh vs. Cong_Suat_Gio_MW

Hệ số tương quan = -0.26 (màu xanh tím nhạt)

-Ý nghĩa:

Có mối quan hệ nghịch biến yếu giữa tiêu thụ điện bình quân và công suất điện gió.

Khi tiêu thụ điện bình quân tăng, thì công suất điện gió có xu hướng giảm nhẹ, và ngược lại.

Tuy nhiên, giá trị này còn xa 0, nên:

Mối liên hệ không đủ mạnh để đưa ra kết luận rõ ràng.

Có thể chỉ là ngẫu nhiên hoặc bị ảnh hưởng bởi yếu tố trung gian như chính sách phát triển năng lượng tái tạo, điều kiện tự nhiên, GDP,…

**3**. Cong_Suat_Gio_MW vs. Cong_Suat_Gio_MW

Hệ số tương quan = 1.00

Giống như trên – biến tương quan tuyệt đối với chính nó.

## 5.Một số chủ đề nâng cao: lựa chọn một trong các chủ thể nâng cao sau để phân tích dữ liệu: Phân tích thành phần chính(PCA)

###5.1.Giới thiệu

+ Khái niệm:
 + Phân tích thành phần chính (PCA) là một kỹ thuật giảm chiều dữ liệu phổ biến trong khai phá dữ liệu, học máy và thống kê. Mục tiêu chính của PCA là chuyển đổi một tập dữ liệu nhiều chiều (có nhiều biến) thành một không gian mới với ít chiều hơn nhưng vẫn giữ được phần lớn thông tin ban đầu.

 + Trong PCA, các chiều mới được gọi là các thành phần chính (Principal Components), là các tổ hợp tuyến tính của những biến gốc và không tương quan với nhau.

+ Tại sao cần PCA?

 + Giảm chiều dữ liệu: Giúp đơn giản hóa mô hình, tăng tốc độ xử lý và trực quan hóa dễ hơn.

 + Loại bỏ nhiễu và dư thừa: Loại bỏ các biến có thông tin lặp lại hoặc ít đóng góp.

 + Trực quan hóa dữ liệu: PCA giúp hiển thị dữ liệu đa chiều trong không gian 2D hoặc 3D để con người dễ quan sát và hiểu.

+ Cách hoạt động của PCA (tóm tắt quy trình)

 + Chuẩn hóa dữ liệu: Đảm bảo các biến có cùng đơn vị (trung bình = 0, độ lệch chuẩn = 1).

 + Tính ma trận hiệp phương sai (Covariance matrix): Đo độ tương quan giữa các biến.

 + Tính toán các giá trị riêng (eigenvalues) và vector riêng (eigenvectors): Tìm các hướng mới của dữ liệu.

 + Chọn các thành phần chính (principal components): Sắp xếp theo lượng phương sai mà mỗi thành phần giải thích.

 + Biến đổi dữ liệu: Chiếu dữ liệu gốc lên các thành phần chính.

+ Ý nghĩa của các thành phần chính

 + PC1 (Thành phần chính 1): Là trục giải thích được nhiều phương sai nhất của dữ liệu.

 + PC2, PC3,...: Là các trục còn lại, mỗi trục giải thích phương sai nhỏ hơn và không tương quan với trục trước đó.

 + Ví dụ: Trong phân tích dữ liệu năng lượng các quốc gia, PC1 có thể đại diện cho quy mô dân số và tiêu thụ năng lượng tổng, còn PC2 có thể thể hiện sự khác biệt về nguồn năng lượng sử dụng (thủy điện, gió, địa nhiệt).

###5.2.Phân tích thành phần chính(PCA)

```r
# Đọc dữ liệu
df <- read.csv("https://raw.githubusercontent.com/Tuananh-zer/file_data.csv/refs/heads/main/danh_sach_tieu_thu_dien_sach.csv", header = TRUE)

# Chuyển dấu phẩy thành dấu chấm nếu có
df$Muc_Tieu_Thu_kWh <- as.numeric(gsub(",", ".", df$Muc_Tieu_Thu_kWh))
df$Dan_so <- as.numeric(gsub(",", ".", df$Dan_so))
df$Tieu_Thu_Binh_Quan_kWh <- as.numeric(gsub(",", ".", df$Tieu_Thu_Binh_Quan_kWh))
df$Cong_Suat_Thuy_Dien_MW <- as.numeric(gsub(",", ".", df$Cong_Suat_Thuy_Dien_MW))
df$Cong_Suat_Gio_MW <- as.numeric(gsub(",", ".", df$Cong_Suat_Gio_MW))
df$Cong_Suat_Dia_Nhiet_MW <- as.numeric(gsub(",", ".", df$Cong_Suat_Dia_Nhiet_MW))

# Lọc các cột numeric và loại bỏ NA
df_numeric <- na.omit(df[, c("Muc_Tieu_Thu_kWh", "Dan_so", "Tieu_Thu_Binh_Quan_kWh",
                             "Cong_Suat_Thuy_Dien_MW", "Cong_Suat_Gio_MW", "Cong_Suat_Dia_Nhiet_MW")])

# Chuẩn hóa dữ liệu
df_scaled <- scale(df_numeric)

# Thực hiện PCA
pca_result <- prcomp(df_scaled, center = TRUE, scale. = TRUE)

# Xem kết quả
summary(pca_result)
biplot(pca_result, scale = 0)
```

Thành phần chính	Tỷ lệ phương sai (%)	Tích lũy (%)

+ PC2	26.46%	57.63%

+ PC3	21.96%	79.60%

+ PC4–PC6	~20.4% còn lại	100%

Kết luận:

+ 3 thành phần đầu tiên (PC1, PC2, PC3) đã giải thích gần 80% tổng phương sai, cho thấy PCA có thể giảm chiều hiệu quả mà vẫn giữ lại thông tin chính.
